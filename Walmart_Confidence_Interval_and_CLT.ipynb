# CI AND CLT

Do males and females spend differently?

```python
# SPLITTING DATASET INTO MALE AND FEMALE
male = df[df['Gender'] == 'M']['Purchase']
female = df[df['Gender'] =='F']['Purchase']
```

```python
# COMPUTING SAMPLE MEAN
male_mean = male.mean().round(2)
female_mean = female.mean().round(2)
print('Male Average Spending =', male_mean, '\nFemale Average Spending =', female_mean)
```

Observation: On average, Males spend around 9437.53 per transaction and
Females spend 8734.57 per transaction (spending per purchase)

```python
# COMPUTE STANDARD ERROR (SE)
import numpy as np

male_se = male.std() / np.sqrt(len(male))
female_se = female.std() / np.sqrt(len(female))
print('Male Standard Error =', male_se, '\nFemale Standard Error =', female_se)
```

```python
# CHOOSE CONFIDENCE LEVEL
z = 1.96
```

```python
# COMPUTE CI (CONFIDENCE INTERVAL)
male_ci = (male_mean - z * male_se, male_mean + z * male_se)
female_ci = (female_mean - z * female_se, female_mean + z * female_se)

print("Male CI:", (round(male_ci[0]), round(male_ci[1])))
print("Female CI:", (round(female_ci[0]), round(female_ci[1])))
```

Observation: The confidence intervals for male and female purchase amounts do not overlap, indicating a statistically significant difference in spending behavior. Male customers have a higher average purchase amount compared to female customers.

```python
# CALCULATING CI USING SCIPY
import numpy as np
from scipy import stats

# Separate data
male = df[df['Gender'] == 'M']['Purchase']
female = df[df['Gender'] == 'F']['Purchase']

# Function for CI
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)  # standard error
    margin = sem * stats.t.ppf((1 + confidence) / 2, len(data)-1)
    return (mean - margin, mean + margin)

# Calculate CI
male_ci = confidence_interval(male)
female_ci = confidence_interval(female)

print("Male CI:", male_ci)
print("Female CI:", female_ci)
```

```python
# COMPUTE CONFIDENCE INTERVAL
married_ci = married_mean - z * married_se, married_mean + z * married_se
unmarried_ci = unmarried_mean - z * unmarried_se, unmarried_mean + z * unmarried_se

print('Married CI =', (round(married_ci[0]), round(married_ci[1])))
print('Unmarried CI =', (round(unmarried_ci[0]), round(unmarried_ci[1])))
```

Observation: The confidence intervals for married and unmarried customers overlap significantly, indicating that marital status does not have a significant impact on purchase behavior.

Does spending differs between age groups?

```python
# GETTING UNIQUE AGE GROUPS
df['Age'].unique()
```

```python
# CALCULATING CI
import numpy as np

z = 1.96  # for 95% CI

age_groups = df['Age'].unique()

for age in age_groups:
    data = df[df['Age'] == age]['Purchase']

    mean = data.mean()
    se = data.std() / np.sqrt(len(data))

    ci_lower = mean - z * se
    ci_upper = mean + z * se

    print(f'Age {age}: Mean = {round(mean, 2)}, CI = ({round(ci_lower)}, {round(ci_upper)})')
```

Observation: The confidence intervals for all age groups overlap significantly, indicating that there is no statistically significant difference in spending behavior across different age groups.
